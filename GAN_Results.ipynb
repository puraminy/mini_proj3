```
import numpy as np
import matplotlib.pyplot as plt
dloss = np.loadtxt('drive/My Drive/GAN/dloss2.txt')
gloss = np.loadtxt('drive/My Drive/GAN/gloss2.txt')
w10=np.where(gloss==-10)
gloss = np.delete(gloss, w10[0])-0
w10=np.where(dloss==-10)
dloss = np.delete(dloss, w10[0])-0
epoch = 70
batch = int(len(dloss)/epoch)
e_gloss = np.zeros(epoch)
e_dloss = np.zeros(epoch)

for i in range (epoch):
  e_gloss[i]=np.mean(gloss[i*batch:(i+1)*batch])
  e_dloss[i]=np.mean(dloss[i*batch:(i+1)*batch])

print(dloss.shape)
print(gloss.shape)

plt.figure(figsize=(10,5))
plt.grid(False)
plt.plot(np.linspace(0,epoch,len(gloss)) ,gloss, 'c',alpha=0.5)
plt.plot(np.linspace(0,epoch,len(dloss)) ,dloss, 'm',alpha=0.5)
plt.plot(e_gloss, 'k', label = 'G_Loss')
plt.plot(e_dloss, 'w', label = 'D_Loss')

plt.legend(loc='upper right')
plt.xlabel('epochs')
plt.ylabel('loss')
```
